### translations with high toxicity

swap translations that have high toxicity with their low-toxicity references, by making the translations the reference and vice versa.

```python
(data.ref_tox < data.trn_tox).mean()
```

```python
data_ref_tox = data.loc[data.ref_tox >= data.trn_tox]
data_ref_tox
```

```python
data_trn_tox = data.loc[data.ref_tox < data.trn_tox]
data_trn_tox
```

```python
data_trn_tox['reference'], data_trn_tox['translation'] = data_trn_tox['translation'], data_trn_tox['reference']
```

```python
data_trn_tox['ref_tox'], data_trn_tox['trn_tox'] = data_trn_tox['trn_tox'], data_trn_tox['ref_tox']
```

```python
data_trn_tox 
```

```python
data_vs = pd.concat([data_ref_tox, data_trn_tox])
data_vs
```

```python
toxic = data_vs.iloc[0:10,1]
detoxified = data_vs.iloc[0:10,2]

toxic_list = list(toxic)
detoxified_list = list(detoxified)

for i in range(10):
    print(toxic_list[i])
    print(detoxified_list[i])
```

```python
data_vs['ref_tox'].plot(kind='kde')
plt.legend(title='ref_tox')
```

```python
data_vs['ref_tox'].mean()
```

```python
data_vs['trn_tox'].plot(kind='kde')
plt.legend(title='trn_tox')
```

```python
data_vs['trn_tox'].mean()
```

## part 4 - data selection

- ref_toxicity >= 0.9
- similarity >= mean of similarity
- lenght_diff <= mean of length_diff

```python
data_final = data_vs.loc[(data_vs.ref_tox >= 0.9) & (data_vs.similarity >= mean_similarity) & (data_vs.lenght_diff <= mean_lenght_diff)]
data_final
```
