```python
import pandas as pd
```

```python
data = pd.read_csv("../data/raw/filtered.tsv", sep="\t")
```

```python
data.head()
```

sorted ascending

```python
data_sorted = data.sort_values(by="ref_tox")
```

```python
data_sorted.head()
```

```python
toxic = data_sorted.iloc[0:10,1]
detoxified = data_sorted.iloc[0:10,2]

toxic_list = list(toxic)
detoxified_list = list(detoxified)

for i in range(10):
    print(toxic_list[i])
    print(detoxified_list[i])
```

sorted in descending order

```python
data_sorted = data.sort_values(by="ref_tox",ascending=False)
```

```python
data_sorted.head()
```

```python
toxic = data_sorted.iloc[0:10,1]
detoxified = data_sorted.iloc[0:10,2]

toxic_list = list(toxic)
detoxified_list = list(detoxified)

for i in range(10):
    print(toxic_list[i])
    print(detoxified_list[i])
```
