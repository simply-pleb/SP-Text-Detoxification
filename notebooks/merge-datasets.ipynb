Merge paranmt selected data with paradetox dataset.

- reference -> toxic
- translation -> neutral1
- neutral2 and neutral3 should be NaN

```python
data_intermit = data_paranmt[['reference', 'translation']]
data_intermit
```

```python
data_intermit = data_intermit.rename(columns={'reference': 'toxic', 'translation': 'neutral1'})
data_intermit
```

```python
data_intermit['neutral2'] = pd.Series([None] * len(data_intermit))
data_intermit['neutral3'] = pd.Series([None] * len(data_intermit))
```

```python
data_result = pd.concat([data_paradetox, data_intermit], ignore_index=True)
data_result
```
